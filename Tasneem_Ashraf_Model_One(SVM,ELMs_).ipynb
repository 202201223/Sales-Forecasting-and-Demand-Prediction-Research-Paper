```python
new_df = pd.read_csv("/content/Super_Store_Data_After_EDA.csv")
new_df
```

```python
y = new_df['Discount']

# Select the feature columns (excluding 'Sales')
X = new_df.drop(columns=['Discount'])
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Create an SVR model
svm_model = SVR(kernel='linear')
svm_model.fit(X_train, y_train)
```

```python
# Predict and evaluate the model
y_pred = svm_model.predict(X_test)
```

```python
# Use regression metrics for evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error (MAE): {mae}')
```

```python
# Prepare data (same as before)
X_encoded = pd.get_dummies(new_df.drop(columns=['Discount']), drop_first=True)
y = new_df['Discount']
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Convert y_train and y_test to NumPy arrays
y_train = y_train.to_numpy().reshape(-1, 1)  # Reshape to 2D array
y_test = y_test.to_numpy().reshape(-1, 1)    # Reshape to 2D array
```

```python
# Create ELM model (1000 hidden neurons)
elm = hpelm.HPELM(X_train.shape[1], 1)
elm.add_neurons(1000, 'sigm')  # Using sigmoid activation
```

```python
# Train the model
elm.train(X_train, y_train, 'train')
```

```python
# Predict using the test set
y_pred = elm.predict(X_test)
```

```python
# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')
```

```python
# Calculate r2
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

```python
# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error (MAE): {mae}')
```
